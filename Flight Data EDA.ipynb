```python
# Loading the train data set
train_df = pd.read_excel("Data_Train.xlsx")
```

```python
# Loading the test data set
test_df = pd.read_excel("Test_set.xlsx")
```

```python
# Appending the data
df = train_df.append(test_df)
```

```python
# See bottom 5 rows of the dataset 
df.tail()
```

```python
# See the data types of the data
df.info()
"""Date of Journey is of object data type we need to convert it into date time """
```

```python
# Converting "Date_of_Journey" to date time object
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'],format='%d/%m/%Y')
```

```python
# Checking the info() of the dataset
df.info()
```

```python
# Creating Month, Day, Year column which is extracted from Date_of_Journey
df['Year'] = df['Date_of_Journey'].dt.year 
df['Month'] =df['Date_of_Journey'].dt.month
df['Day'] = df['Date_of_Journey'].dt.day
```

```python
# Checking the head of the dataset
df.head()
```

```python
# Checking the unique values in Day, Month, Year Column
print("Unique Values in Year column:")
print(df['Year'].unique()) # This shows we have data only of 1 year i.e 2019
print("Unique Values in Month column:")
print(df['Month'].unique()) # This shows we have data for Mar, April, May & June
# We can  perform same operation using string split method
```

```python
## Checking the info of newly created variables
df.info()
```

```python
# Drop column : "Date_of_Journey"
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
# Head of the dataset
df.head()
```

```python
## Arrival Time
# Removed the date where available
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x :x.split(" ")[0])
```

```python
# Head() of dataset
df.head()
```

```python
# Checking the null values
df.isnull().sum()
```

```python
## Extracting the Arrical hours and minutes
df['Arrival_Hrs'] = df['Arrival_Time'].apply(lambda x : int(x.split(":")[0]))
df['Arrival_Min'] = df['Arrival_Time'].apply(lambda x : int(x.split(":")[1]))
# Astype works with series data type
```

```python
# Head()
df.head()
```

```python
# Info()
df.info()
```

```python
# Dropping the arrival time column
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
# Head()
df.head(2)
```

```python
## Fixing Departure time
```

```python
df['Dep_Hrs'] = df['Dep_Time'].apply(lambda x: int(x.split(":")[0]))
df['Dep_Min'] = df['Dep_Time'].apply(lambda x: int(x.split(":")[1]))
df.head(2)
```

```python
# Dropping Dep_Time
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
## Route
```

```python
# Checking first 10 rows of Route coulumns
df['Route'].head(10)
```

```python
# Total Stops
df['Total_Stops'].unique()
# Value_counts()
df['Total_Stops'].value_counts()
```

```python
# Checking the details where Total_Stops column has null value
df[df['Total_Stops'].isna()]
```

```python
# Done the Mapping
"""{'non-stop':0,'1 stop':1,'2 stops':2,"3 stops":3,"4 stops":4,'nan':1}"""
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,"3 stops":3,"4 stops":4,'nan':1})
```

```python
# Drop the Route columns
df.drop('Route',axis=1,inplace=True)
```

```python
#Checking head()
df.head()
```

```python
df['Duration_min'] = df['Duration'].apply(lambda x:str(x.split(" ")[0]).replace('h',""))
```

```python
df = df[df['Duration_min']!='5m']
```

```python
df.head()
```

```python
# Converting it into "int"
df['Duration_min'] = df['Duration_min'].astype(int)
# Converting hours into minutes
df['Duration_min'] =df['Duration_min']*60
```

```python
df.head()
```

```python
df['Duration_mins'] = df['Duration'].apply(lambda x:x.split(" ")[-1])
```

```python
df = df[~df['Duration_mins'].str.contains('h')]
```

```python
df.head()
```

```python
df['Duration_mins'] = df['Duration_mins'].str.replace("m","").astype(int)
```

```python
df.info()
```

```python
# Adding both mins
df['Duration_Mins'] = df['Duration_min']+df['Duration_mins']
```

```python
# First few rows of Dataset
df.head()
```

```python
# Droping Unwanted columns at once
df.drop(["Duration","Duration_mins","Duration_min"],axis=1,inplace = True)
```

```python
## Handling Airline columns

# Get the unique airline names
df['Airline'].unique()
```

```python
# Feature Engineering Importing package for the samez
from sklearn.preprocessing import LabelEncoder
```

```python
# Creating object for LabelEncoder
l_en = LabelEncoder()
```

```python
# Doing for all categorical variables
df['Airline'] = l_en.fit_transform(df['Airline'])
df['Source'] = l_en.fit_transform(df['Source'])
df['Destination'] = l_en.fit_transform(df['Destination'])
df['Additional_Info'] = l_en.fit_transform(df['Additional_Info'])
```
